```python
def delta(A, n):
    return (A - A.shift(n)).fillna(
        0
    )

def wash(A):
    temp = A.fillna(0)
    temp = temp.replace([np.inf, -np.inf], 0)
    return temp

def max_(A, B):
    return pd.concat([A,B]).max(level=0)


def max_ASquare(A1,B1,C1,A2,B2,C2):
    a = max_(A1**2, B1)
    a = max_(a, C1)
    b = max_(A2**2, B2)
    b = max_(b, C2)
    return a, b

def cumAGreaterBtwopfive(A,B,C):
    a = A > B  * 2.5
    b = a & a.shift(1) & a.shift(2)
    c = b * C
    return c

def delay(A, n):
    temp = A.shift(n)
    temp = temp.fillna(0)
    return temp

@jit(nopython = True)
def pass_resistance_ask(thres_arr, trueprice_arr):
    looked1, looked2, looked3 = 0, 0, 0
    toReturn = np.zeros(len(thres_arr))  
    for i in range(len(thres_arr)):
        thres = thres_arr[i]
        price = trueprice_arr[i]
        #print('threshold is ', max(looked1, looked2, looked3))
        #print('price is ', price)
        #print('thres is ', thres)
        #print('------------------')  
        if thres != 0:
            looked2, looked3 = looked1, looked2
            looked1 = thres
        elif price >= max(looked1, looked2, looked3):
            continue
        elif price < max(looked1, looked2, looked3):
            toReturn[i] = 1
            looked1, looked2, looked3 = 0, 0, 0
    return toReturn

@jit(nopython = True)
def pass_resistance_bid(thres_arr, trueprice_arr):
    looked1, looked2, looked3 = 10000, 10000, 10000
    toReturn = np.zeros(len(thres_arr))  
    for i in range(len(thres_arr)):
        thres = thres_arr[i]
        price = trueprice_arr[i]
        #print('threshold is ', min(looked1, looked2, looked3))
        #print('price is ', price)
        #print('thres is ', thres)
        #print('------------------')  
        if thres != 0:
            looked2, looked3 = looked1, looked2
            looked1 = thres
        elif price <= min(looked1, looked2, looked3):
            continue
        elif price > min(looked1, looked2, looked3):
            toReturn[i] = 1
            looked1, looked2, looked3 = 10000, 10000, 10000
    return toReturn

file_dic
```

```python
'''
Multi-Level Order-Flow Imbalance in a Limit Order Book by Ken et al. Page 6.
'''

def deltaW(BidPrice, BidVolume):
    case1 = BidPrice > BidPrice.shift(1)
    case2 = BidPrice == BidPrice.shift(1)
    case3 = BidPrice < BidPrice.shift(1)
    toReturn = case1 * BidVolume + case2 * (BidVolume - BidVolume.shift(1).fillna(0)) + case3 * -1 * BidVolume.shift(1)
    return toReturn

def deltaV(AskPrice, AskVolume):
    case1 = AskPrice > AskPrice.shift(1)
    case2 = AskPrice == AskPrice.shift(1)
    case3 = AskPrice < AskPrice.shift(1)
    toReturn = case1 * -1 * AskVolume + case2 * (AskVolume - AskVolume.shift(1).fillna(0)) + case3 * AskVolume.shift(1)
    return toReturn

def deltaW1(BidPrice, BidVolume):
    case1 = BidPrice < BidPrice.shift(1)
    case2 = BidPrice == BidPrice.shift(1)
    case3 = BidPrice < BidPrice.shift(1)
    toReturn = case1 * -1 * BidVolume + case2 * (BidVolume - BidVolume.shift(1).fillna(0)) + case3 * -1 * BidVolume.shift(1)
    return toReturn

def deltaV1(AskPrice, AskVolume):
    case1 = AskPrice < AskPrice.shift(1)
    case2 = AskPrice == AskPrice.shift(1)
    case3 = AskPrice < AskPrice.shift(1)
    toReturn = case1 * AskVolume + case2 * (AskVolume - AskVolume.shift(1).fillna(0)) + case3 * AskVolume.shift(1)
    return toReturn

def OFI(BidPrice, BidVolume, AskPrice, AskVolume):
    case1 = BidPrice >= BidPrice.shift(1)
    case2 = BidPrice <= BidPrice.shift(1)
    case3 = AskPrice >= AskPrice.shift(1)
    case4 = AskPrice <= AskPrice.shift(1)
    toReturn = case1 * BidVolume - case2 * BidVolume.shift(1).fillna(0) - case4 * AskVolume - case3 * AskVolume.shift(1)
    return toReturn

def oneMinOFI(OFI):
    toReturn = np.zeros(len(OFI))
    for i in range(int(np.floor(len(OFI) / 20))):
        sum_ = 0
        for j in range(20):
            sum_ += OFI[i*20+j]
        toReturn[i*20:(i+1)*20] = sum_
    toReturn[(i+1)*20:] = sum(OFI[(i+1)*20:])
    return toReturn

def autoregOFI(OFI, df, param=100):
    toReturn = np.zeros(len(OFI))
    for i in range(int(len(OFI)/param)):
        reg = LinearRegression().fit([[1, i] for i in OFI[i*param:(i+1)*param].to_numpy()], df['MidReturn'][i*param:(i+1)*param].fillna(0))
        toReturn[(i+1)*param:(i+2)*param] = reg.coef_[1]*OFI[(i+1)*param:(i+2)*param]
    return toReturn

def Press(BidPrice, AskPrice, BidVolume, AskVolume, TruePrice):
    case1 = BidPrice - TruePrice
    case2 =  AskPrice - TruePrice
    p = case1*BidVolume - case2*AskVolume
    return p

def VolumeRatio(tick, df):
    toReturn = np.zeros(len(df['Volume']))
    for i in range(int(np.floor(len(df['Volume']) / tick))):
        vol = df['TrueVolume'][i*tick:(i+1)*tick]
        mr = df['MidReturn'][i*tick:(i+1)*tick]
        vr = (sum((mr>0)*vol)+np.mean(vol)/2) / (sum((mr<0)*vol+np.mean(vol)/2))
        toReturn[(i+1)*tick:(i+2)*tick] = vr
    return toReturn

def shift1(arr, num, fill_value=np.nan):
    arr = np.roll(arr,num)
    if num < 0:
        arr[num:] = fill_value
    elif num > 0:
        arr[:num] = fill_value
    return arr
```

```python
dirty = './Data_588000_trunc_VWAP'
for i in range(len(file_dic)):
    df = pd.read_pickle(file_dic[i])
    df = df.dropna()
    df['time'] = df.index
    df = df[~df['time'].duplicated(keep='first')]

    df['ImBalance1'] = ( (df['BidVolume1']) -  (df['AskVolume1'])) / ( (df['BidVolume1']) +  (df['AskVolume1']))
    #df['ImBalance1'] = df['ImBalance1']
    df['ImBalance2'] = ( (df['BidVolume2']) -  (df['AskVolume2'])) / ( (df['BidVolume2']) +  (df['AskVolume2']))
    df['ImBalance3'] = ( (df['BidVolume3']) -  (df['AskVolume3'])) / ( (df['BidVolume3']) +  (df['AskVolume3']))
    df['ImBalance4'] = ( (df['BidVolume4']) -  (df['AskVolume4'])) / ( (df['BidVolume4']) +  (df['AskVolume4']))
    df['ImBalance5'] = ( (df['BidVolume5']) -  (df['AskVolume5'])) / ( (df['BidVolume5']) +  (df['AskVolume5']))
    df['ImBalance10'] = ( (df['BidVolume10']) -  (df['AskVolume10'])) / ( (df['BidVolume10']) +  (df['AskVolume10']))
    df['ImBalance1v'] = df['ImBalance1'] - df['ImBalance1'].shift(1).fillna(0)
    df['ImBalance2v'] = df['ImBalance2'] - df['ImBalance2'].shift(1).fillna(0)
    df['ImBalance3v'] = df['ImBalance3'] - df['ImBalance3'].shift(1).fillna(0)
    FirstHfHR = np.zeros(len(df['ImBalance1']))
    FirstHfHR[:600] = 1
    df['FirstHfHR'] = FirstHfHR
    df['OFIa'] = (deltaW(df['BidPrice1'], df['BidVolume1']) - deltaV(df['AskPrice1'], df['AskVolume1'])).fillna(0)
    df['OFI2a'] = (deltaW(df['BidPrice2'], df['BidVolume2']) - deltaV(df['AskPrice2'], df['AskVolume2'])).fillna(0)
    df['OFI3a'] = (deltaW(df['BidPrice3'], df['BidVolume3']) - deltaV(df['AskPrice3'], df['AskVolume3'])).fillna(0)
    
    df['OFIb'] = (deltaW1(df['BidPrice1'], df['BidVolume1']) - deltaV1(df['AskPrice1'], df['AskVolume1'])).fillna(0)
    df['OFI2b'] = (deltaW1(df['BidPrice2'], df['BidVolume2']) - deltaV1(df['AskPrice2'], df['AskVolume2'])).fillna(0)
    df['OFI3b'] = (deltaW1(df['BidPrice3'], df['BidVolume3']) - deltaV1(df['AskPrice3'], df['AskVolume3'])).fillna(0)
    
    df['OFI4'] = (deltaW(df['BidPrice4'], df['BidVolume4']) - deltaV(df['AskPrice4'], df['AskVolume4'])).fillna(0)
    df['OFI5'] = (deltaW(df['BidPrice5'], df['BidVolume5']) - deltaV(df['AskPrice5'], df['AskVolume5'])).fillna(0)
    df['OFI10'] = (deltaW(df['BidPrice10'], df['BidVolume10']) - deltaV(df['AskPrice10'], df['AskVolume10'])).fillna(0)
    df['OFI1v'] = df['OFIa']  - df['OFIa'].shift(1).fillna(0)
    #df['OFI2v'] = df['OFIV2']  - df['OFIV2'].shift(1).fillna(0)
    #df['OFI3v'] = df['OFIV3']  - df['OFIV3'].shift(1).fillna(0)
    #df['OFI1'] = shift1(oneMinOFI(df['OFI']), 20)
    df['OFIV'] = ((deltaW(df['BidPrice1'], df['BidVolume1']) - deltaV(df['AskPrice1'], df['AskVolume1'])).fillna(0))/(np.abs(deltaW(df['BidPrice1'], df['BidVolume1'])) + np.abs(deltaV(df['AskPrice1'], df['AskVolume1'])))
    df['OFIV2'] = ((deltaW(df['BidPrice2'], df['BidVolume2']) - deltaV(df['AskPrice2'], df['AskVolume2'])).fillna(0))/(np.abs(deltaW(df['BidPrice2'], df['BidVolume2'])) + np.abs(deltaV(df['AskPrice2'], df['AskVolume2'])))
    df['OFIV3'] = ((deltaW(df['BidPrice3'], df['BidVolume3']) - deltaV(df['AskPrice3'], df['AskVolume3'])).fillna(0))/(np.abs(deltaW(df['BidPrice3'], df['BidVolume3'])) + np.abs(deltaV(df['AskPrice3'], df['AskVolume3'])))
    
    s5b = (df['BidPrice5'] - df['BidPrice1']) / (df['BidVolume5'] - df['BidVolume1']) 
    s5a = (df['AskPrice5'] - df['AskPrice1']) / (df['AskVolume5'] - df['AskVolume1']) 
    df['slope5'] = (s5b - s5a) / (np.abs(s5b) + np.abs(s5a))

    s10b = (df['BidPrice10'] - df['BidPrice1']) / (df['BidVolume10'] - df['BidVolume1']) 
    s10a = (df['AskPrice10'] - df['AskPrice1']) / (df['AskVolume10'] - df['AskVolume1']) 
    df['slope10'] = (s10b - s10a) / (np.abs(s10b) + np.abs(s10a))
    
    df['TrueVolume'] = df['Volume'] - df['Volume'].shift(1)
    df['TrueVolume'] = df['TrueVolume'].fillna(0)
    df['TruePrice'] = df[['BidPrice1', 'AskPrice1']].mean(axis =1)
    df['MeanRev2'] = (df['TruePrice'] - df['TruePrice'].rolling(20).mean()).fillna(0)
    df['MeanRev1'] = (df['TruePrice'] - df['TruePrice'].rolling(10).mean()).fillna(0)
    #df['Press'] = Press(df['BidPrice2'], df['AskPrice2'], df['BidVolume2'], df['AskVolume2'],  df['TruePrice'])
    df['Mid20VWAP'] = (df['TrueVolume'] * df['TruePrice']).rolling(20).sum() / df['TrueVolume'].rolling(20).sum()
    df['SampleY'] = (df['Mid20VWAP'].shift(-20) - df['TruePrice']) / df['TruePrice']
    
    
    
    df['MidReturn'] = df['TruePrice'].pct_change(1).shift(-1).fillna(0)
    #df['VolumeRatio1'] = VolumeRatio(20, df)
    #df['VolumeRatio2'] = VolumeRatio(40, df)
    #df['AuroRegOFI'] = autoregOFI(df['OFI'], df, 100)
    df['Momentum1'] = df['MidReturn'].shift(1)
    df['Momentum2'] = df['MidReturn'].shift(2)
    df['Momentum3'] = df['MidReturn'].shift(3)
    df['MidReturnMag'] = df['MidReturn'].apply(lambda x: 10**7*x**2 if x >= 0 else -10**7*x**2)
    thres_arrA = cumAGreaterBtwopfive(df['AskVolume1'],df['BidVolume1'],df['TruePrice'])
    df['overA'] = pass_resistance_ask(thres_arrA.to_numpy(), df['TruePrice'].to_numpy())
    thres_arrB = cumAGreaterBtwopfive(df['BidVolume1'],df['AskVolume1'],df['TruePrice'])
    df['overB'] = pass_resistance_bid(thres_arrB.to_numpy(), df['TruePrice'].to_numpy())

    df.to_pickle(dirty + '/' + file_dic[i][20:30] + '.txt')
    print(file_dic[i][20:30])
    
```
